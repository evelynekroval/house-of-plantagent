```python
import requests # Must be the python package
import os
from dotenv import load_dotenv

load_dotenv()
```

```python
API_NINJAS_KEY = os.getenv("API_NINJAS_KEY")

# And then it looks like instead of printing it, you do a safety check...
if not API_NINJAS_KEY:
    raise RuntimeError(
        "API_NINJAS_KEY not found in .env. "
        "Did you create .env from .env.example and add your key?"
    )
print(f"Key loaded successfully (first 10 chars): {API_NINJAS_KEY[:10]}...")
```

```python
import requests # Must be the python package
from .env import NINJAS_API_KEY, RECIPE_SOURCE_URL

headers = {
    "X-Api-Key": NINJAS_API_KEY
}

# Make a GET request to an API endpoint

response = requests.get(RECIPE_SOURCE_URL), headers = headers

# Check if the request succeeded
if response.status_code == 200:
    data = response.json() # Parse the successful JSON into a Python dict
    print(data)
else:
    print(f"Error: {response.status_code}")
```
